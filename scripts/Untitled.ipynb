```python
import pandas as pd

jobs = {}

for el in soup.find_all("li", class_="expanded top"):
    job = el.find_next("a").text
    jobs[job] = []
    
    for li in el.find_all("li", class_="leaf"):
        jobs[job].append(li.find_next("a").text)
        
all_df = pd.DataFrame()    
for k, v in jobs.items():
    df = pd.DataFrame({"category": [k] * len(v), "job": v})
    all_df = pd.concat([all_df, df])

all_df.to_csv("jobs.csv", index=False)
```
